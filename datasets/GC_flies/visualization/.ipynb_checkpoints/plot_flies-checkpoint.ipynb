```python
import os
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20.0, 10.0))
margin = 0.01

data = np.load('../npy/fly_data.npy', allow_pickle=True)

for number in range(len(data)):

    px1 = data[number][0][2]
    py1 = data[number][0][3]
    px2 = data[number][1][2]
    py2 = data[number][1][3]
    px3 = data[number][2][2]
    py3 = data[number][2][3]
    px4 = data[number][3][2]
    py4 = data[number][3][3]
    px5 = data[number][4][2]
    py5 = data[number][4][3]
    px6 = data[number][5][2]
    py6 = data[number][5][3]
    px7 = data[number][6][2]
    py7 = data[number][6][3]
    px8 = data[number][7][2]
    py8 = data[number][7][3]

    x_min = np.min([np.min(px1), np.min(px2), np.min(px3), np.min(px4), np.min(px5), np.min(px6), np.min(px7), np.min(px8)])
    x_max = np.max([np.max(px1), np.max(px2), np.max(px3), np.max(px4), np.max(px5), np.max(px6), np.max(px7), np.max(px8)])
    y_min = np.min([np.min(py1), np.min(py2), np.min(py3), np.min(py4), np.min(py5), np.min(py6), np.min(py7), np.min(py8)])
    y_max = np.max([np.max(py1), np.max(py2), np.max(py3), np.max(py4), np.max(py5), np.max(py6), np.max(py7), np.max(py8)])

    range_max = np.array([x_max - x_min, y_max - y_min]).max() * 0.5
    x_mid = (x_max + x_min) * 0.5
    y_mid = (y_max + y_min) * 0.5

    # plot
    plt.subplot(1,2, number+1)
    plt.plot(px1, py1)
    plt.plot(px2, py2)
    plt.plot(px3, py3)
    plt.plot(px4, py4)
    plt.plot(px5, py5)
    plt.plot(px6, py6)
    plt.plot(px7, py7)
    plt.plot(px8, py8)

    plt.xlim(x_mid - range_max - margin, x_mid + range_max + margin)
    plt.ylim(y_mid - range_max - margin, y_mid + range_max + margin)
    plt.xlabel("x-position [m]")
    plt.ylabel("y-position [m]")
```
