```python
%matplotlib notebook
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation


data = np.load('../npy/flies_data.npy', allow_pickle=True)

for number in range(1, 3):
    data_number = number - 1

    px1 = data[data_number][0][2]
    py1 = data[data_number][0][3]
    px2 = data[data_number][1][2]
    py2 = data[data_number][1][3]
    px3 = data[data_number][2][2]
    py3 = data[data_number][2][3]
    px4 = data[data_number][3][2]
    py4 = data[data_number][3][3]
    px5 = data[data_number][4][2]
    py5 = data[data_number][4][3]
    px6 = data[data_number][5][2]
    py6 = data[data_number][5][3]
    px7 = data[data_number][6][2]
    py7 = data[data_number][6][3]
    px8 = data[data_number][7][2]
    py8 = data[data_number][7][3]

    x_min = np.min([np.min(px1), np.min(px2)])
    x_max = np.max([np.max(px1), np.max(px2)])
    y_min = np.min([np.min(py1), np.min(py2)])
    y_max = np.max([np.max(py1), np.max(py2)])

    range_max = np.array([x_max - x_min, y_max - y_min]).max() * 0.5
    x_mid = (x_max + x_min) * 0.5
    y_mid = (y_max + y_min) * 0.5


    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_aspect('equal')
    margin = 0.01
    samplerate = 30
    time_width = 3

    def update_func(i):

        ax.clear()
        ax.plot(px1[i-time_width:i], py1[i-time_width:i])
        ax.plot(px2[i-time_width:i], py2[i-time_width:i])
        ax.plot(px3[i-time_width:i], py3[i-time_width:i])
        ax.plot(px4[i-time_width:i], py4[i-time_width:i])
        ax.plot(px5[i-time_width:i], py5[i-time_width:i])
        ax.plot(px6[i-time_width:i], py6[i-time_width:i])
        ax.plot(px7[i-time_width:i], py7[i-time_width:i])
        ax.plot(px8[i-time_width:i], py8[i-time_width:i])

        ax.set_xlim(x_mid - range_max - margin, x_mid + range_max + margin)
        ax.set_ylim(y_mid - range_max - margin, y_mid + range_max + margin)
        ax.set_xlabel("x-position [m]")
        ax.set_ylabel("y-position [m]")

        number_frame = i
        ax.set_title('                                  fly_' +  "%02d" % number + '               Frame: ' + "%03d" % number_frame )


    # save movie
    ani = animation.FuncAnimation(fig, update_func, frames=len(px1), interval=1000/samplerate, repeat=True)
    name = 'flies_' +  "%02d" % number + '.mp4'
    filename = '../video/flies/' + name
    ani.save(filename, writer="ffmpeg")
    plt.show()
```
