```python
%matplotlib notebook
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from mpl_toolkits.mplot3d import Axes3D


data = np.load('../npy/peregrine_data.npy', allow_pickle=True)


for number in range(1, 23):
    data_number = number - 1

    px1 = data[data_number][0][3]
    py1 = data[data_number][0][4]
    pz1 = data[data_number][0][5]
    px2 = data[data_number][1][3]
    py2 = data[data_number][1][4]
    pz2 = data[data_number][1][5]

    x_min = np.min([np.min(px1), np.min(px2)])
    x_max = np.max([np.max(px1), np.max(px2)])
    y_min = np.min([np.min(py1), np.min(py2)])
    y_max = np.max([np.max(py1), np.max(py2)])
    z_min = np.min([np.min(pz1), np.min(pz2)])
    z_max = np.max([np.max(pz1), np.max(pz2)]) 
    range_max = np.array([x_max - x_min, y_max - y_min, z_max - z_min]).max() * 0.5
    x_mid = (x_max + x_min) * 0.5
    y_mid = (y_max + y_min) * 0.5
    z_mid = (z_max + z_min) * 0.5


    # 3D
    fig = plt.figure()
    ax = Axes3D(fig)
    margin = 10
    samplerate = 5
    time_width = 3

    def update_func(i):

        ax.clear()

        ax.plot(px1[i-time_width:i], py1[i-time_width:i], pz1[i-time_width:i])
        ax.plot(px2[i-time_width:i], py2[i-time_width:i], pz2[i-time_width:i])

        ax.set_xlim(x_mid - range_max, x_mid + range_max)
        ax.set_ylim(y_mid - range_max, y_mid + range_max)
        ax.set_zlim(z_mid - range_max, z_mid + range_max)
        ax.set_xlabel("x-position [m]")
        ax.set_ylabel("y-position [m]")
        ax.set_zlabel("z-position [m]")

        number_frame = i
        ax.set_title('                                peregrine_' +  "%02d" % number + '               Frame: ' + "%03d" % number_frame)


    # save movie
    ani = animation.FuncAnimation(fig, update_func, frames=len(px1), interval=1000/samplerate, repeat=True)
    name = 'peregrine_3d_' +  "%02d" % number + '.mp4'
    filename = '../video/peregrine/' + name
    ani.save(filename, writer="ffmpeg")
    plt.show()
```

```python
%matplotlib notebook
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation


data = np.load('../npy/peregrine_data.npy', allow_pickle=True)


for number in range(1, 23):
    data_number = number - 1

    px1 = data[data_number][0][3]
    py1 = data[data_number][0][4]
    pz1 = data[data_number][0][5]
    px2 = data[data_number][1][3]
    py2 = data[data_number][1][4]
    pz2 = data[data_number][1][5]

    x_min = np.min([np.min(px1), np.min(px2)])
    x_max = np.max([np.max(px1), np.max(px2)])
    y_min = np.min([np.min(py1), np.min(py2)])
    y_max = np.max([np.max(py1), np.max(py2)])
    z_min = np.min([np.min(pz1), np.min(pz2)])
    z_max = np.max([np.max(pz1), np.max(pz2)]) 
    range_max = np.array([x_max - x_min, y_max - y_min, z_max - z_min]).max() * 0.5
    x_mid = (x_max + x_min) * 0.5
    y_mid = (y_max + y_min) * 0.5
    z_mid = (z_max + z_min) * 0.5


    # 2D
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_aspect('equal')
    margin = 10
    samplerate = 5
    time_width = 3

    def update_func(i):

        ax.clear()
        
        ax.plot(px1[i-time_width:i], py1[i-time_width:i])
        ax.plot(px2[i-time_width:i], py2[i-time_width:i])

        ax.set_xlim(x_mid - range_max - margin, x_mid + range_max + margin)
        ax.set_ylim(y_mid - range_max - margin, y_mid + range_max + margin)
        ax.set_xlabel("x-position [m]")
        ax.set_ylabel("y-position [m]")

        number_frame = i
        ax.set_title('                                peregrine_' +  "%02d" % number + '               Frame: ' + "%03d" % number_frame )


    # save movie
    ani = animation.FuncAnimation(fig, update_func, frames=len(px1), interval=1000/samplerate, repeat=True)
    name = 'peregrine_2d_' +  "%02d" % number + '.mp4'
    filename = '../video/peregrine/' + name
    ani.save(filename, writer="ffmpeg")
    plt.show()
```
