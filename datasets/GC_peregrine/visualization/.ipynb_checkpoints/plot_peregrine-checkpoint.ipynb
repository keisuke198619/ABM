```python
import os
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(20.0, 30.0))
margin = 10

data = np.load('../npy/peregrine_data.npy', allow_pickle=True)
print(np.array(data).shape)

for number in range(len(data)):

    px1 = data[number][0][3]
    py1 = data[number][0][4]
    pz1 = data[number][0][5]
    px2 = data[number][1][3]
    py2 = data[number][1][4]
    pz2 = data[number][1][5]

    x_min = np.min([np.min(px1), np.min(px2)])
    x_max = np.max([np.max(px1), np.max(px2)])
    y_min = np.min([np.min(py1), np.min(py2)])
    y_max = np.max([np.max(py1), np.max(py2)])
    z_min = np.min([np.min(pz1), np.min(pz2)])
    z_max = np.max([np.max(pz1), np.max(pz2)])
    
    range_max = np.array([x_max - x_min, y_max - y_min, z_max - z_min]).max() * 0.5
    x_mid = (x_max + x_min) * 0.5
    y_mid = (y_max + y_min) * 0.5
    z_mid = (z_max + z_min) * 0.5

    # plot 3D
    ax = fig.add_subplot(6, 4, number+1, projection='3d')
    
    ax.plot(px1, py1, pz1)
    ax.plot(px2, py2, pz2)

    ax.set_xlim(x_mid - range_max, x_mid + range_max)
    ax.set_ylim(y_mid - range_max, y_mid + range_max)
    ax.set_zlim(z_mid - range_max, z_mid + range_max)
    ax.set_xlabel("x-position [m]")
    ax.set_ylabel("y-position [m]")
    ax.set_zlabel("z-position [m]")
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20.0, 30.0))
margin = 10

data = np.load('../npy/peregrine_data.npy', allow_pickle=True)
print(np.array(data).shape)

for number in range(len(data)):

    px1 = data[number][0][3]
    py1 = data[number][0][4]
    pz1 = data[number][0][5]
    px2 = data[number][1][3]
    py2 = data[number][1][4]
    pz2 = data[number][1][5]

    x_min = np.min([np.min(px1), np.min(px2)])
    x_max = np.max([np.max(px1), np.max(px2)])
    y_min = np.min([np.min(py1), np.min(py2)])
    y_max = np.max([np.max(py1), np.max(py2)])
    z_min = np.min([np.min(pz1), np.min(pz2)])
    z_max = np.max([np.max(pz1), np.max(pz2)])
    
    range_max = np.array([x_max - x_min, y_max - y_min, z_max - z_min]).max() * 0.5
    x_mid = (x_max + x_min) * 0.5
    y_mid = (y_max + y_min) * 0.5
    z_mid = (z_max + z_min) * 0.5

    # plot 2D
    plt.subplot(6,4, number+1)
    plt.plot(px1, py1)
    plt.plot(px2, py2)

    plt.xlim(x_mid - range_max - margin, x_mid + range_max + margin)
    plt.ylim(y_mid - range_max - margin, y_mid + range_max + margin)
    plt.xlabel("x-position [m]")
    plt.ylabel("y-position [m]")
```
